```python
import numpy as np
import scipy as sp
from scipy import stats
from scipy.stats import norm
from scipy.stats import poisson
import matplotlib.pyplot as plt
```

```python
np.random.seed(312)
data = np.random.randn(200)
plt.hist(data, density = 1)
xr = np.linspace(-3, 3, 100)

mean, stdev = sp.stats.norm.fit(data)
plt.plot(xr, norm.pdf(xr, loc = mean, scale = stdev), label = "norm")

plt.plot(xr, norm.pdf(xr, loc = 0, scale = 1))
plt.xlabel("Some Data")
plt.ylabel("Frequency")
```

```python
mean, stdev = sp.stats.norm.fit(data)
mean, stdev
```

```python
l = 1
data = np.random.poisson(l, 200)
xr = np.arange(0, 5, 1)
xr_continous = np.linspace(0, 5, 100)
plt.hist(data, density = 1)
mean, stdev = sp.stats.norm.fit(data)

plt.plot(xr_continous, norm.pdf(xr_continous, loc = mean, scale = stdev), label = "norm")

plt.plot(xr, poisson.pmf(xr, mu = 1), label = "true")
plt.xlabel("Some Data")
plt.ylabel("Count")
```
